# This Script takes the EV hub articles, parses all the EV related and creates a sitemap for them. The sitemap can then be uploaded to 

```python
import requests
import xml.etree.ElementTree as ET

```

```python
# Fetch the sitemap XML
url = "https://guide.autoscout24.ch/post-sitemap.xml"
response = requests.get(url)
response_content = response.content

# Parse the XML
root = ET.fromstring(response_content)
namespace = {'ns': 'http://www.sitemaps.org/schemas/sitemap/0.9'}

# Extract all links
links = [elem.text for elem in root.findall('ns:url/ns:loc', namespace)]
links
```

```python

prefixes = ['https://guide.autoscout24.ch/fr/mobilite-electrique/', 'https://guide.autoscout24.ch/it/mobilita-elettrica/', 'https://guide.autoscout24.ch/de/elektromobilitaet/']
```

```python
# Filter out the links that don't start with the prefixes
filtered_links = [link for link in links if link.startswith(tuple(prefixes))]
```

```python
#save the links to a txt file
with open('site_links.txt', 'w') as f:
    for item in filtered_links:
        f.write("%s\n" % item)
```

```python
sitemap_template = """<?xml version="1.0" encoding="UTF-8"?>
<urlset xmlns="http://www.sitemaps.org/schemas/sitemap/0.9">
{urls}
</urlset>"""

url_template = """
    <url>
        <loc>{loc}</loc>
    </url>"""

urls = "\n".join([url_template.format(loc=link) for link in links])

sitemap_content = sitemap_template.format(urls=urls)

# Save the sitemap to a file
with open('sitemap.xml', 'w') as file:
    file.write(sitemap_content)
```
